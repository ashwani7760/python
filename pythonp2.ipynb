## generator is the other way to operating iterator

```python
def square(n):
    for i in range(n):
        return  i**2
```

```python
square(2)
```

```python
#generator
def square(n):
    for i in range(n):
        yield  i**2
```

```python
k = square(3)
```

```python
type(k)
```

```python
for i in k:
    print(i)
```

```python
next(k)
```

```python
print(k)
```

```python
# genertaor has multiple yield but function doesnt have multiple return 
```

```python
def new():
    n=1
    print("one")
    yield n

    n = n+1
    print("two")
    yield n

    n = n+1
    print("three")
    yield n
```

```python
for i in new():
    print(i)
```
