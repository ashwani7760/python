## level configuation

```python
import logging

logging.basicConfig(level = logging.DEBUG)

def add(m,n):
    return m+n
def sub(m,n):
    return m-n
def mul(m,n):
    return m*n
def div(m,n):
    return m/n

a=20
b=10

addr = add(a,b)
logging.debug(addr)
subr = sub(a,b)
logging.info(subr)
mulr = mul(a,b)
logging.warning(mulr)
divr = div(a,b)
logging.error(divr)
```

## format

```python
import logging
logging.basicConfig(filename = "nameformat.log",level  = logging.DEBUG,format = "%(asctime)s -  %(levelname)s - %(message)s")

def namecheck(name):
    if len(name)<2:
        logging.debug("check len of name")
        return "invalid name"
    elif name.isspace():
        logging.debug("name hasspace")
        return "invalid name"
    elif name.isspace():
        logging.debug("checking for name is alphabet")
        return "valid name"
    elif name.replace(" ","").isalpha():
        logging.debug("checking for name")
        return "valid name"
    else:
        logging.debug("all check fails")
        return "invalid name"
```

```python
namecheck("ash wani")
```

```python
namecheck("    ")
```

```python
import logging
logging.basicConfig(filename = "task1.log",level  = logging.DEBUG,format = "%(asctime)s -  %(levelname)s - %(message)s")

def s(*args):
    try:
        for i in args:
            if type(i)==int:
                pass
            else:
                raise ValueError("not integer")
    except Exception as e:
        print(e)
    else:
        logging.debug(f"{args} - sum {sum(args)}")
```

```python
import logging
# stream handler to display info on console
logger = logging.getLogger(__name__)

#filehandler

f_handler = logging.FileHandler
f_handler.setLevel(logging.WARNING) # set level for file handler
f_format = logging.Formatter("%(asctime)s -  %(levelname)s - %(message)s") # define format 
f_handler.setFormatter(f_format)   # set format to file handler

# streamhadler

s_handler = logging.streamHandler("asc.log") #create stream handler
s_handler.setLevel(logging.DEBUG) # set level for stream handler
s_format = logging.Formatter("%(asctime)s -  %(levelname)s - %(message)s") # define format 
s_handler.setFormatter(s_format)   # set format to stream handler

# add handler to the loggers
logger.addHandler(f_handler)
logger.addHandler(s_handler)


logger.warning("this is warning")
logger.error("this is error")
```
