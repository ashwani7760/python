## level configuation

```python
import logging

logging.basicConfig(level = logging.DEBUG)

def add(m,n):
    return m+n
def sub(m,n):
    return m-n
def mul(m,n):
    return m*n
def div(m,n):
    return m/n

a=20
b=10

addr = add(a,b)
logging.debug(addr)
subr = sub(a,b)
logging.info(subr)
mulr = mul(a,b)
logging.warning(mulr)
divr = div(a,b)
logging.error(divr)
```

## format

```python
import logging
logging.basicConfig(filename = "nameformat.log",level  = logging.DEBUG,format = "%(asctime)s -  %(levelname)s - %(message)s")

def namecheck(name):
    if len(name)<2:
        logging.debug("check len of name")
        return "invalid name"
    elif name.isspace():
        logging.debug("name hasspace")
        return "invalid name"
    elif name.isspace():
        logging.debug("checking for name is alphabet")
        return "valid name"
    elif name.replace(" ","").isalpha():
        logging.debug("checking for name")
        return "valid name"
    else:
        logging.debug("all check fails")
        return "invalid name"
```

```python
namecheck("ash wani")
```

```python
namecheck("    ")
```
