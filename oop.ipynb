```python
#class and object
# create class
class car:
    #function / method
    def information(self):
        print(f"{self}\n ------------- ")
        print(f"brand : {self.brand} \n model : {self.model} \n year : {self.year} \n price : {self.price} \n fueltype : {self.fueltype}")
```

```python
# create object or instance
c1  = car()
```

```python
# attribute or properties of instance/object
c1.brand = "suzuki"
c1.model = "swift"
c1.year= 2025
c1.price = 850000
c1.fueltype = "cng"
```

```python
c1.information()
```

```python
class circle:

    def __init__(self,radius):
        self.radius = radius

    def area(self):
        return 3.14*self.radius*self.radius
    def circumference(self):
        return 2*3.14*self.radius
```

```python
c1 = circle(7)
```

```python
c1.area()
```

```python
c1.circumference()
```

```python
class bankAccount :
    def __init__(self,acn,name,balance=2000):
        self.acn = acn
        self.name = name
        self.balance = balance
    def display(self):
        print("acn    name    balance")
        print(self.acn,self.name,self.balance)
    def deposit(self,rs):
        self.balance = self.balance +rs
        print("transaction complete your balance : ",self.balance)
    def withdraw(self,rs):
        if rs>self.balance:
            print("you have insufficient balance")
            print("transaction complete your balance:",self.balance)
        else:
            self.balance = self.balance -rs
            print("transaction complete your balance :",self.balance)
```

```python
a1 = bankAccount(140101511111,"ashwani")
```

```python
a1.display()
```

```python
a1.deposit(20000)
```

```python
a1.display()
```

```python
a1.withdraw(22000)
```

```python
a1.display()
```

```python
a1.withdraw(22000)
```
