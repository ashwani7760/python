```python
class circle:

    def __init__(self,radius):
        self.radius = radius

    def area(self):
        return 3.14*self.radius*self.radius
    def circumference(self):
        return 2*3.14*self.radius
```

```python
class bankAccount :
    def __init__(self,acn,name,balance=2000):
        self.acn = acn
        self.name = name
        self.balance = balance
    def display(self):
        print("acn    name    balance")
        print(self.acn,self.name,self.balance)
    def deposit(self,rs):
        self.balance = self.balance +rs
        print("transaction complete your balance : ",self.balance)
    def withdraw(self,rs):
        if rs>self.balance:
            print("you have insufficient balance")
            print("transaction complete your balance:",self.balance)
        else:
            self.balance = self.balance -rs
            print("transaction complete your balance :",self.balance)
```

```python
class dictionary_parser:
    try:
        def __init__(d,dictionary):
            d.dictionary = dictionary
    except ValueError as e:
        print("please enter a dictionary")

    def key(d):
        l = []
        for k in d.dictionary:
            l.append(k)
        return l

    def value(d):
        l = []
        for v in d.dictionary:
            l.append(d.dictionary[v])
        return l
    def updat(d,k,v):
        try:
            d.dictionary.update({k:v})
            print("update sucessful")
        except Exception as e:
            print(e)
    def check(d):
        a = input("enter key : ")
        for i in d.dictionary:
            if i == a.lower():
                print(a," : ",d.dictionary[i])
            else :
                print("not found in dictionary")
                break
    def show(d):
        for i in d.dictionary:
            print(i," : ",d.dictionary[i])
```

```python
l1  = {"oppo":"phone","canon":"camera","hp":["laptop","printer"],}
```

```python
dic = dictionary_parser(l1)
```

```python
dic.key()
```

```python
dic.value()
```

```python
dic.updat("apple","seb")
```

```python
dic.show()
```

```python
dic.check()
```

```python
class file:
    def __init__(self):
        self.fname= input("enter file name: ")
    def read(self):
        try:
            with open(self.fname) as f:
                return f.read()
        except:
            print("entered file is not found ")
    def write(self):
        text = input("enter text to add :")
        with open(self.fname, "a") as f:
            f.write(text)
        print("write successfully")
```

```python
f1 = file()
```

```python
f1.read()
```

```python
f1.write()
```

```python
class BonusCalculator:
    def __init__(self,empid,emprating):
        self.empid = empid
        self.emprating = emprating
        self._empmail = "abc@gmail.com"
        self.__bonusforratingA = "70%"
        self.__bonusforratingB = "60%"
        self.__bonusforratingC = "40%"

    def bonuscalculator(self):
        if self.emprating == "A":
            bonus = self.__bonusforratingA
            return bonus
        elif self.emprating == "B":
            bonus = self.__bonusforratingB
            return bonus
        elif self.emprating == "C":
            bonus = self.__bonusforratingC
            return bonus
        else:
            print("no bonus")
```
