```python
# make list of string from other list
def tes(a):
    ''' this is a function thats help you to seperate string from list'''
    l = []
    for i in a:
        if type(i) == str:
            l.append(i)
    return l
```

```python
tes(l)
```

```python
def ind(e,l):
    '''give index of element in given list'''
    for i in range(len(l)):
        if l[i] == e:
            return i
            break
    else:
        print("search element is not in list")
```

```python
ind(46,l)
```
